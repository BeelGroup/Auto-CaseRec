```python
import time
results = {'itemknn': [], 'userknn': [], 'popular': [], 'bprmf': [], 'random': []}
time_vals = {}

start = time.time()
for i in range(5):
    
    ItemKNN(train_file="datasets/ml-latest-small/folds/"+str(i)+"/train.dat", test_file="datasets/ml-latest-small/folds/"+str(i)+"/test.dat",
                                output_file="datasets/ml-latest-small/folds/"+str(i)+"/result.dat", rank_length=10).compute()
    
    results['itemknn'].append(ItemRecommendationEvaluation(n_ranks=[10]).evaluate_with_files("datasets/ml-latest-small/folds/"+str(i)+"/result.dat",
                                                                       "datasets/ml-latest-small/folds/"+str(i)+"/test.dat")['MAP@10'])

end = time.time()
time_vals['itemknn'] = float(end-start)/60

start = time.time()
for i in range(5):
    
    BprMF(train_file="datasets/ml-latest-small/folds/"+str(i)+"/train.dat", test_file="datasets/ml-latest-small/folds/"+str(i)+"/test.dat",
                                output_file="datasets/ml-latest-small/folds/"+str(i)+"/result.dat", rank_length=10).compute()
    
    results['bprmf'].append(ItemRecommendationEvaluation(n_ranks=[10]).evaluate_with_files("datasets/ml-latest-small/folds/"+str(i)+"/result.dat",
                                                                       "datasets/ml-latest-small/folds/"+str(i)+"/test.dat")['MAP@10'])

end = time.time()
time_vals['bprmf'] = float(end-start)/60

start = time.time()
for i in range(5):
    
    MostPopular(train_file="datasets/ml-latest-small/folds/"+str(i)+"/train.dat", test_file="datasets/ml-latest-small/folds/"+str(i)+"/test.dat",
                                output_file="datasets/ml-latest-small/folds/"+str(i)+"/result.dat", rank_length=10).compute()
    
    results['popular'].append(ItemRecommendationEvaluation(n_ranks=[10]).evaluate_with_files("datasets/ml-latest-small/folds/"+str(i)+"/result.dat",
                                                                       "datasets/ml-latest-small/folds/"+str(i)+"/test.dat")['MAP@10'])

end = time.time()
time_vals['popular'] = float(end-start)/60

start = time.time()
for i in range(5):
    
    UserKNN(train_file="datasets/ml-latest-small/folds/"+str(i)+"/train.dat", test_file="datasets/ml-latest-small/folds/"+str(i)+"/test.dat",
                                output_file="datasets/ml-latest-small/folds/"+str(i)+"/result.dat", rank_length=10).compute()
    
    results['userknn'].append(ItemRecommendationEvaluation(n_ranks=[10]).evaluate_with_files("datasets/ml-latest-small/folds/"+str(i)+"/result.dat",
                                                                       "datasets/ml-latest-small/folds/"+str(i)+"/test.dat")['MAP@10'])

end = time.time()
time_vals['userknn'] = float(end-start)/60

start = time.time()
for i in range(5):
    
    RandomRec(train_file="datasets/ml-latest-small/folds/"+str(i)+"/train.dat", test_file="datasets/ml-latest-small/folds/"+str(i)+"/test.dat",
                                output_file="datasets/ml-latest-small/folds/"+str(i)+"/result.dat", rank_length=10).compute()
    
    results['random'].append(ItemRecommendationEvaluation(n_ranks=[10]).evaluate_with_files("datasets/ml-latest-small/folds/"+str(i)+"/result.dat",
                                                                       "datasets/ml-latest-small/folds/"+str(i)+"/test.dat")['MAP@10'])

end = time.time()
time_vals['random'] = float(end-start)/60
    
 
```
