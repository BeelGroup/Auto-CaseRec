```python
import os
import csv

def write_results(i, results):
    keys = results[0].keys()
    with open('experiments/item_recommendation/lastfm_test'+str(i)+'.csv', 'w', newline='') as output_file:
        dict_writer = csv.DictWriter(output_file, keys)
        dict_writer.writeheader()
        dict_writer.writerows(results)       
            
```

```python
kwargs = {'datapath': "datasets/last.fm/"+'user_artists.dat', 
                            'predictor':'item', 
                            'eval_metric':'MAP', 
                            'eval_rank':10, 
                            'early_stop_split':2, 
                            'algo':'tpe', 
                            'max_evals': 50, 
                            'cross_validate':True, 
                            'test_percentage':0.2,
                            'sep_read':"\t", 
                            'sep_write':"\t",
                            'n_splits':5, 
                            'dir_folds': "datasets/last.fm/", 
                            'header':1, 
                            'names':[0,1,2]}
```

```python
bests = []
resultss = []

for i in range(1,6,1):
    print("Test : " + str(i))
    myEstimator = AutoEstimator(**kwargs)
    best, results = myEstimator.fit()
    write_results(i, results)
    bests.append(best)
    resultss.append(results)
```

```python
import time
default_results = {'itemknn': [], 'userknn': [], 'popular': [], 'bprmf': [], 'random': []}
time_vals = {'itemknn': [], 'userknn': [], 'popular': [], 'bprmf': [], 'random': []}



for n in range(5):
    start = time.time()
    for i in range(5):
    
        ItemKNN(train_file="datasets/ml-latest-small/folds/"+str(i)+"/train.dat", test_file="datasets/ml-latest-small/folds/"+str(i)+"/test.dat",
                                output_file="datasets/ml-latest-small/folds/"+str(i)+"/result.dat", rank_length=10).compute()
    
        default_results['itemknn'].append(ItemRecommendationEvaluation().evaluate_with_files("datasets/ml-latest-small/folds/"+str(i)+"/result.dat",
                                                                       "datasets/ml-latest-small/folds/"+str(i)+"/test.dat")['MAP@10'])

    end = time.time()
    time_vals['itemknn'].append(float(end-start)/60)

    start = time.time()
    for i in range(5):
    
        BprMF(train_file="datasets/ml-latest-small/folds/"+str(i)+"/train.dat", test_file="datasets/ml-latest-small/folds/"+str(i)+"/test.dat",
                                output_file="datasets/ml-latest-small/folds/"+str(i)+"/result.dat", rank_length=10).compute()
    
        default_results['bprmf'].append(ItemRecommendationEvaluation(n_ranks=[10]).evaluate_with_files("datasets/ml-latest-small/folds/"+str(i)+"/result.dat",
                                                                       "datasets/ml-latest-small/folds/"+str(i)+"/test.dat")['MAP@10'])

    end = time.time()
    time_vals['bprmf'].append(float(end-start)/60)

    start = time.time()
    for i in range(5):
        
        MostPopular(train_file="datasets/ml-latest-small/folds/"+str(i)+"/train.dat", test_file="datasets/ml-latest-small/folds/"+str(i)+"/test.dat",
                                output_file="datasets/ml-latest-small/folds/"+str(i)+"/result.dat", rank_length=10).compute()
    
        default_results['popular'].append(ItemRecommendationEvaluation(n_ranks=[10]).evaluate_with_files("datasets/ml-latest-small/folds/"+str(i)+"/result.dat",
                                                                       "datasets/ml-latest-small/folds/"+str(i)+"/test.dat")['MAP@10'])

    end = time.time()
    time_vals['popular'].append(float(end-start)/60)

    start = time.time()
    for i in range(5):
    
        UserKNN(train_file="datasets/ml-latest-small/folds/"+str(i)+"/train.dat", test_file="datasets/ml-latest-small/folds/"+str(i)+"/test.dat",
                                output_file="datasets/ml-latest-small/folds/"+str(i)+"/result.dat", rank_length=10).compute()
    
        default_results['userknn'].append(ItemRecommendationEvaluation(n_ranks=[10]).evaluate_with_files("datasets/ml-latest-small/folds/"+str(i)+"/result.dat",
                                                                       "datasets/ml-latest-small/folds/"+str(i)+"/test.dat")['MAP@10'])

    end = time.time()
    time_vals['userknn'].append(float(end-start)/60)

    start = time.time()
    for i in range(5):
    
        RandomRec(train_file="datasets/ml-latest-small/folds/"+str(i)+"/train.dat", test_file="datasets/ml-latest-small/folds/"+str(i)+"/test.dat",
                                output_file="datasets/ml-latest-small/folds/"+str(i)+"/result.dat", rank_length=10).compute()
    
        default_results['random'].append(ItemRecommendationEvaluation(n_ranks=[10]).evaluate_with_files("datasets/ml-latest-small/folds/"+str(i)+"/result.dat",
                                                                       "datasets/ml-latest-small/folds/"+str(i)+"/test.dat")['MAP@10'])

    end = time.time()
    time_vals['random'].append(float(end-start)/60)   
 
```
